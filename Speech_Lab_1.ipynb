```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import librosa
import librosa.display
import scipy.signal

filename = "LJ001-0017.wav"
signal, sample_rate = librosa.load(filename, sr=None)

num_samples = len(signal)
duration = num_samples / sample_rate
print(f"Sample Rate: {sample_rate} Hz")
print(f"Number of Samples: {num_samples}")
print(f"Duration: {duration:.2f} seconds")

plt.figure(figsize=(10, 4))
librosa.display.waveshow(signal, sr=sample_rate)
plt.title("Speech Signal Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

sliced_signal = signal[:2 * sample_rate]
plt.figure(figsize=(10, 4))
librosa.display.waveshow(sliced_signal, sr=sample_rate)
plt.title("Sliced Signal (First 2 seconds)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

normalized_signal = signal / np.max(np.abs(signal))
plt.figure(figsize=(10, 4))
librosa.display.waveshow(normalized_signal, sr=sample_rate)
plt.title("Normalized Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

amplified_signal = 2 * signal
plt.figure(figsize=(10, 4))
librosa.display.waveshow(amplified_signal, sr=sample_rate)
plt.title("Amplified Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

deamplified_signal = 0.5 * signal
plt.figure(figsize=(10, 4))
librosa.display.waveshow(deamplified_signal, sr=sample_rate)
plt.title("De-Amplified Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

upsampled_signal = scipy.signal.resample(signal, 2 * num_samples)
plt.figure(figsize=(10, 4))
librosa.display.waveshow(upsampled_signal, sr=2 * sample_rate)
plt.title("Up-Sampled Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

downsampled_signal = signal[::2]
plt.figure(figsize=(10, 4))
librosa.display.waveshow(downsampled_signal, sr=sample_rate // 2)
plt.title("Down-Sampled Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

sf.write("sliced_signal.wav", sliced_signal, sample_rate)
sf.write("normalized_signal.wav", normalized_signal, sample_rate)
sf.write("amplified_signal.wav", amplified_signal, sample_rate)
sf.write("deamplified_signal.wav", deamplified_signal, sample_rate)
sf.write("upsampled_signal.wav", upsampled_signal, 2 * sample_rate)
sf.write("downsampled_signal.wav", downsampled_signal, sample_rate // 2)


```
